```python
import requests as reqs
import pandas as pd
import numpy as np

url = "https://api.punkapi.com/v2/beers?page="
sess = reqs.session()

resp = sess.get(f"{url}1&per_page=80").json()
df = pd.DataFrame(resp)

cont = True
index = 2
while cont:
    resp = sess.get(f"{url}{index}&per_page=80").json()
    index += 1
    if len(resp) == 0:
        cont = False
    df = pd.concat([pd.DataFrame(resp), df])
    # df = df.append(pd.DataFrame(resp))
df.shape
```

```python
df = df.sort_values(by="ibu", ascending=False)
df.head()
```

```python
@np.vectorize
def isvalid(x):
    return not np.isnan(x) and 80 < x and x < 95

df = df[isvalid(df["attenuation_level"])]
df.head()
```

```python
@np.vectorize
def m_temp(input: str) -> int:
    return dict(input)["mash_temp"][0]["temp"]["value"]

df["mash_temp"] = m_temp(df["method"])
df.head()
```

```python
list_of_foods = ["chicken", "cheese", "chocolate"]
beer_map = {}

@np.vectorize
def match_pairing(foods: list, food: str):
    out = False
    for f in foods:
        out = out or food in f.lower()
    return out

for f in list_of_foods:
    beer_map[f] = df[match_pairing(df["food_pairing"].values, f)]["name"].values.tolist()

beer_map
```

```python
beer_intersection = [beer for beer in beer_map["chicken"] if (beer in beer_map["chocolate"]) and (beer in beer_map["cheese"])]
if beer_intersection == []:
    print("non of them goes with all three foods")
else:
    print(beer_intersection)

# beer_map["cheese"]
```
