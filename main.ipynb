```python
import requests as reqs
import pandas as pd
import numpy as np

url = "https://api.punkapi.com/v2/beers?page="
sess = reqs.session()

resp = sess.get(f"{url}1&per_page=80").json()
df = pd.DataFrame(resp)

cont = True
index = 2
while cont:
    resp = sess.get(f"{url}{index}&per_page=80").json()
    index += 1
    if len(resp) == 0:
        cont = False
    df = pd.concat([pd.DataFrame(resp), df], axis=0)
    # df = df.append(pd.DataFrame(resp))
df.set_index(keys="id", inplace=True)
df.head()
```

```python
df = df[(~df["attenuation_level"].isna()) & (df["attenuation_level"]>80) & (df["attenuation_level"]<95)]
df.head()
```

```python
df["mash_temp"] = df["method"].apply(lambda method: dict(method)["mash_temp"][0]["temp"]["value"])
df.head()
```

```python
list_of_foods = ["chicken", "cheese", "chocolate"]
beer_map = {}

@np.vectorize
def match_pairing(foods: list, food: str):
    out = False
    for f in foods:
        out = out or food in f.lower()
    return out

for f in list_of_foods:
    beer_map[f] = df[match_pairing(df["food_pairing"].values, f)]["name"].values.tolist()

beer_map
```

```python
beer_intersection = set(beer_map["chicken"]) & set(beer_map["chocolate"]) & set(beer_map["cheese"])
if beer_intersection:
    print(beer_intersection)
else:
    print("non of them goes with all three foods")
```
